```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('Iris.csv')
df.head()
```

```python
df.describe()
```

```python
#visualize the whole dataset
sns.pairplot(df , hue='Id')
```

```python
#seprate features and target
data = df.values

X = data[:,0:4]
Y = data[:,5]
print(X)
```

```python
#split the data to train and test dataset
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size=0.2)
print(X_test)
```

```python
#support vector machine algorithm
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train , Y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test , prediction1)*100)
for i in range(len(prediction1)):
  print(Y_test[i],prediction1[i])
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train , Y_train)
```

```python
prediction2 = model_LR.predict(X_test)
#calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test , prediction2)*100)
for i in range(len(prediction2)):
  print(Y_test[i],prediction2[i])
```

```python
# A detailed classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test , prediction2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =  KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, Y_train)
```

```python
X_new = np.array([[5,2.9,1,0.2]])
print("X_new_shape: {}".format(X_new.shape))
```

```python
prediction3 = knn.predict(X_new)
print("Prediction:{}".format(prediction3))
#the result is 0 so that it is iris setosa.  
```

```python
X_new = np.array([[51, 7.0, 3.2, 4.7],[4,2,3.8,1.1],[6.3,3.3,6.0,2.5]])
# prediction of the species from the input vector
prediction = model_LR.predict(X_new)
print("Prediction of species: {}".format(prediction))
```
